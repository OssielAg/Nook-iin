# Multilayer Heterostructure
The following sections demonstrate the procedure to obtain the Primitive Cell (PC) for a heterostructure composed of multiple layers. The systems evaluated here are purely theoretical, with no purpose other than to illustrate the functionality of Nook'iin.

In each example, the system will be created, its primitive cell will be computed ensuring that the deformation in each layer remains below a given ε, the corresponding data table and real-space representation will be displayed, the result will be exported as a POSCAR file, and finally, its diffraction pattern will be shown.

The materials used in these examples are:

- $\beta$-GeSe

- CdS

- hBN

- WS₂

- WSe₂

- Black Phosphorene

The rotation angles of each layer will be specified individually in each example.

### Loading System and Lattice Creation

```python
import sys
import os

src_path = os.path.abspath(os.path.join(os.getcwd(), '..', '..', 'src'))
if src_path not in sys.path:
    sys.path.append(src_path)

from System import *
```

```python
VASP_Files = os.path.abspath(os.path.join(os.getcwd(), '..', '..', 'VASP_Files'))
bGeSe = importLattice(VASP_Files+"/GeSe_Beta")
CdS = importLattice(VASP_Files+"/CdS")
hBN = importLattice(VASP_Files+"/hBN")
WS2 = importLattice(VASP_Files+"/WS2")
WSe2 = importLattice(VASP_Files+"/WSe2")
bP = importLattice(VASP_Files+"/Ph")
```

### System: Black Phosphorene-hBN$(3.25^\circ)$-CdS$(8.76^\circ)$-hBN$(12.62^\circ)$

```python
S1 = System([bP,hBN.mRot(3.25),CdS.mRot(8.76),hBN.mRot(12.62)])
S1_op = S1.generateSuperCell(RoS=20,eps=0.01,prntRes=False,showTable=True)
S1_op.SuperRed.showme(t=1)
S1_op.exportPC()
#S1_op.diffractionPattern(border=1)
```

### System: WS2-WSe2$(1.25^\circ)$-CdS$(-3.48^\circ)$-BlackPhosphorene$(15.83^\circ)$-$\beta$_GeSe$(8.67^\circ)$

```python
S2 = System([WS2,WSe2.mRot(1.25),CdS.mRot(-3.48),bP.mRot(15.83),bGeSe.mRot(8.67)])
S2_op = S2.generateSuperCell(RoS=25,eps=0.02,prntRes=False,showTable=True)
S2_op.SuperRed.showme(t=1)
S2_op.exportPC()
#S2_op.diffractionPattern(border=1)
```

### System: $\beta$_GeSe-WS2$(12.53^\circ)$-hBN$(5.28^\circ)$-CdS$(11.68^\circ)$

```python
S3 = System([bGeSe,WS2.mRot(12.53),hBN.mRot(5.28),CdS.mRot(11.68)])
S3_op = S3.generateSuperCell(RoS=20,eps=0.01,prntRes=False,showTable=True)
S3_op.SuperRed.showme(t=1)
S3_op.exportPC()
#S3_op.diffractionPattern(border=1)
```

### System CdS-bGeSe$(3.25^\circ)$-hBN$(6.74^\circ)$-WS2$(11.42^\circ)$-WSe2$(9.74^\circ)$-BlackPhosphorene$(-3.25^\circ)$-CdS$(-8.1^\circ)$-bGeSe$(13.69^\circ)$-hBN$(1.24^\circ)$-WS2$(-5.74^\circ)$-WSe2$(3.68^\circ)$-BlackPhosphorene

```python
S4 = System([CdS,
             bGeSe.mRot(3.25),
             hBN.mRot(6.74),
             WS2.mRot(13.34),
             WSe2.mRot(9.74),
             bP.mRot(-4.37),
             CdS.mRot(-8.1),
             bGeSe.mRot(13.42),
             hBN.mRot(1.24),
             WS2.mRot(-5.74),
             WSe2.mRot(3.68),
             bP])
S4_op = S4.generateSuperCell(RoS=20,eps=0.02,prntRes=False,showTable=True)
S4_op.SuperRed.showme(t=0.1)
S4_op.exportPC()
#S3_op.diffractionPattern(border=1)
```

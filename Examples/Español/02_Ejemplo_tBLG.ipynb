# Grafeno bicapa rotado

A continuación se mostrará el procedimiento para obtener la PC para una bicapas de grafeno con distintas rotaciones.

Las rotaciones relativas entre las capas de grafeno que conformarán cada sistema serán en ángulos no-conmensurables. Estos ángulos serán los siguientes:

-  $2.54^\circ$
-  $5.63^\circ$
- $14.21^\circ$
- $16.18^\circ$
- $23.85^\circ$

*La explicación paso a paso del calculo está descrito en el archivo "01_Ejemplo_Completo"

### Carga de System y creación de grafenos

```python
import sys
import os

src_path = os.path.abspath(os.path.join(os.getcwd(), '..', '..', 'src'))
if src_path not in sys.path:
    sys.path.append(src_path)

from System import *
```

```python
c1 = graphene()
c2 = graphene()
c1.showPC()
c2.showPC()
```

### tBLG con 2.54$^\circ$

```python
t = 2.54
tBLG = System([c1,c2.mRot(t)])
TM = tBLG.ejecuta(n=20,e=0.01)
_=tBLG.describeTM(TM)
tBLG_op,D = tBLG.optimize_system(TM)
tBLG.exportPC(PCname=f"tBLG({t})")
tBLG.diffractionPattern(border=1)
```

### tBLG 5.63$^\circ$

```python
t = 5.63
tBLG = System([c1,c2.mRot(t)])
tBLG_op = tBLG.generateSuperCell(RoS=20,eps=0.01,showTable=True)
tBLG_op.exportPC(PCname=f"tBLG({t})")
tBLG_op.diffractionPattern(border=1)
```

### tBLG 14.21$^\circ$

```python
t = 14.21
tBLG = System([c1,c2.mRot(t)])
tBLG_op = tBLG.generateSuperCell(RoS=20,eps=0.01,showTable=True)
tBLG_op.exportPC(PCname=f"tBLG({t})")
tBLG_op.diffractionPattern(border=1)
```
